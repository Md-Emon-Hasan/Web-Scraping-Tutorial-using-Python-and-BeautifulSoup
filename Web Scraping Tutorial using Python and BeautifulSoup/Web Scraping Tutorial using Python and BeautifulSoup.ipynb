```python
# step 0: install all the requirements
import requests
from bs4 import BeautifulSoup
url = 'https://codewithharry.com'

# step 1: Get the HTML
r = requests.get(url)
htmlContent = r.content
# print(htmlContent)

# step 2: Parse the HTML
soup = BeautifulSoup(htmlContent,'html.parser')
# print(soup)
print(soup.prettify)
print('**********')

# step 3: Tree traverse the HTML
# 1. Tag
# 2. NavigableString
# 3. BeautifulSoup
# 4. Comment
title = soup.title
print(title)
print('**********')
# 1. Tag
print(type(title))
print('**********')
# 2. NavigableString
print(type(title.string))
print('**********')
# 3. BeautifulSoup
print(type(soup))
```

```python
import requests
from bs4 import BeautifulSoup
url = 'https://codewithharry.com'

# Get the title from the page
print('Get the title from the page:')
title = soup.title
print(title)
print('**********')

# Get all the paragraph from the page
print('Get all the paragraph from the page:')
paras = soup.find_all('p')
print(paras)
print('**********')

# get all anchor tags from the page
print('get all anchor tags from the page:')
anchors = soup.find_all('a')
print(anchors)
print('**********')

# Find a class
print('Find a class:')
print(soup.find('p'))
print('**********')

# Find classes of any element in the HTML page
print('Find classes of any element in the HTML page:')
print(soup.find('p')['class'])
print('**********')

# Get id of any element in the HTML page
print('Get id of any element in the HTML page:')
print(soup.find('p'),['id'])
print('**********')

# Find all the elements with class lead
print('Find all the elements with class lead:')
print(soup.find_all('p',class_='lead'))
print('**********')

# Get the text from the tags/soup
print('Get the text from the tags/soup:')
print(soup.find('p').get_text())
print('**********')

# Get all text from page
print('Get all text from page:')
print(soup.get_text())
print('**********')

# Get all the links on the page
print('Get all the links on the page:')
for link in anchors:
    print(link.get('href'))
```

```python
import requests
from bs4 import BeautifulSoup
url = 'https://codewithharry.com'

# Get all the anchors tags from the page
anchors = soup.find_all('a')
all_links = set()
# Get all the links on the page
print('Get all the links on the page:')
for link in anchors:
    if(link != '#'):
        link = 'https://codewithharry.com' + link.get('href')
        all_links.add(link)
        print(link)
print('**********')

# Get all the anchors tags from the page
anchors = soup.find_all('a')
all_links = set()
# Get all the links on the page
print('Get all the links on the page:')
for link in anchors:
    if(link.get('href') != '#'):
        linkText = 'https://codewithharry' + link.get('href')
        all_links.add(link)
        print(linkText)
```
